```python
from pymongo.mongo_client import MongoClient
from pymongo.server_api import ServerApi
import psycopg2

# Establishment Postgres Connection    
# Replace these variables with your actual database connection details
dbname = 'postgres'
user = 'dap'
password = 'dap'
host = 'localhost'  # often 'localhost'
port = '5432'  # often 5432

connection = psycopg2.connect(
    dbname=dbname,
    user=user,
    password=password,
    host=host,
    port=port
)

# Create a cursor object to interact with the database
cursor = connection.cursor()

# Execute a SELECT query to retrieve some records from the table
cursor.execute("SELECT * FROM MODELS LIMIT 5 ")  # Limiting to 5 records for demonstration purposes

# Fetch all the rows returned by the SELECT query
rows = cursor.fetchall()

# Print the retrieved rows
for row in rows:
    print(row)
    
# SQL query to get all column names from a table
query = """
    SELECT column_name 
    FROM information_schema.columns 
    WHERE table_name = 'MODELS';
"""

# Execute the query
cursor.execute(query)

# Fetch all the rows returned by the query
rows = cursor.fetchall()

# Extract column names from the result
column_names = [row[0] for row in rows]

# Print the column names
print("Column Names:")
for column_name in column_names:
    print(column_name)

# Close the cursor and connection
cursor.close()
connection.close()
```
